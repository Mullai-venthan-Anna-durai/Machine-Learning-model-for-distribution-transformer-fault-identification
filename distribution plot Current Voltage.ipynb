```python
transformer_failure_val = [6,4,4,4,3,2,28,27,13,9]
transformer_failure = ["Electrical Connection", "Overload","Foreign objects","Moisture","Line Disturbance","Stress / Fatigue","Electrical Disturbances","others","Lightning","Insulation"]
```

```python
plt.axis("equal")
plt.pie(transformer_failure_val,labels=transformer_failure, radius = 2,autopct='%0.0f%%', shadow=True, explode=[0,0.3,0,0.5,0.5,0.6,0,0,0,0.4])
plt.show()
```
